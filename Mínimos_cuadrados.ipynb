## Explicación del Método

Dados $(x_0,y_0),(x_1,y_1),...,(x_n,y_n)$, la optimización por mínimos cuadrados busca estimar los parámetros de $b_0$, $b_1$ que cumplen:
\begin{equation}
\hat{y}=\hat{b}_0\vec{x}+\hat{b}_1
\end{equation}
La solucion para b es:
\begin{equation}
\hat{b}=(X^{T}X)^{-1}X^{T}Y
\end{equation}
Donde  $X=[[x0,1],[x1,1],...,[xn,1]]$ es una matriz de tamaño $n\times 2$, $\hat{b}=[[b_0],[b_1]]$ un array de tamaño $2\times 1$

## Implementación del Método

* seleccionamos un conjunto de puntos semi-aleatorios

```python
n=100
x=np.linspace(-10,10,n)
y=2*(x+5*np.random.rand(n))+1*np.random.rand(n)
```

```python
plt.figure(figsize=(8,8))
#Gráfica:
plt.scatter(x,y,color='tab:olive',label="Datos")
plt.xlabel("$x$")#Etiquetas de los ejes en latex
plt.ylabel("$y$") #Etiquetas de los ejes en latex
plt.legend(loc=(1.1/1,2./4)) #Para colocar la leyenda
```

Aplicamos la fórmula para hallar b0 y b1
\begin{equation}
\hat{b}=(X^{T}X)^{-1}X^{T}Y
\end{equation}

```python
X=np.ones((len(x),2))
X[:,0]=x
x_t_x=np.dot(X.T,X)
# x_col=x.reshape(100,1)#numpy no distingue entre un vector y su transpuesto por eso se cambian sus dimensionesasi
# x_row=x.reshape(1,100)#numpy no distingue entre un vector y su transpuesto
```

```python
beta=np.dot(np.linalg.inv(x_t_x),np.dot(X.T,y))
print(beta)
```

## Predicciones:

```python
y_hat=np.dot(X,beta)
```

```python
plt.figure(figsize=(8,8))
#Gráfica:
plt.scatter(x,y,color='tab:olive',label="Datos")
plt.plot(x,y_hat,color='tab:red',label=r"$y={:3f}x+{:3f}$".format(beta[0],beta[1]))
plt.xlabel("$x$")#Etiquetas de los ejes en latex
plt.ylabel("$y$") #Etiquetas de los ejes en latex
plt.legend(loc=(1.1/1,2./4)) #Para colocar la leyenda
```
